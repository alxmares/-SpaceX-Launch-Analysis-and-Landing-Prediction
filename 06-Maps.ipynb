```python
import folium
import pandas as pd

# Import folium MarkerCluster plugin
from folium.plugins import MarkerCluster
# Import folium MousePosition plugin
from folium.plugins import MousePosition
# Import folium DivIcon plugin
from folium.features import DivIcon
```

```python
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
spacex_df = pd.read_csv(URL)
spacex_df.head()
```

```python
# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
coastline_lat = 28.5632
coastline_lon = -80.568

launch_site_lat = spacex_df.loc[0, 'Lat']
launch_site_lon = spacex_df.loc[0, 'Long']
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
```

```python
distance_marker = folium.Marker(
    location=[coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance_coastline),
    )
)
site_map.add_child(distance_marker)

# Mostrar el mapa
site_map
```

```python
coordinates = [[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]]
lines = folium.PolyLine(locations=coordinates, weight=1)
site_map.add_child(lines)

# Mostrar el mapa
site_map
```

```python
points_of_interest = {
    'Coastline': (28.5632, -80.568),
    'Closest City': (28.388, -80.6057),
    'Railway': (28.5711, -80.5854),
    'Highway': (28.5624, -80.5706)
}
```

```python
# Calcular y marcar distancias a los puntos de interés
for name, (poi_lat, poi_lon) in points_of_interest.items():
    launch_site_lat = launch_sites_df.loc[0, 'Lat']
    launch_site_lon = launch_sites_df.loc[0, 'Long']
    distance = calculate_distance(launch_site_lat, launch_site_lon, poi_lat, poi_lon)

    # Crear y añadir un marcador en el punto de interés
    distance_marker = folium.Marker(
        location=[poi_lat, poi_lon],
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance),
        )
    )
    site_map.add_child(distance_marker)
    
    # Añadir una línea que conecte el sitio de lanzamiento con el punto de interés
    coordinates = [[launch_site_lat, launch_site_lon], [poi_lat, poi_lon]]
    lines = folium.PolyLine(locations=coordinates, weight=1)
    site_map.add_child(lines)

# Mostrar el mapa
site_map
```
